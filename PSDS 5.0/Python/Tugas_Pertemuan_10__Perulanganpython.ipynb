#**Looping/Perulangan**

Python menyediakan dua statement untuk melakukan proses perulangan yaitu `for` and `while`. Diantara kedua statement ini, secara umum `for` lebih banyak digunakan daripada `while`.

##For Loop

Perulangan for digunakan untuk melakukan iterasi di dalam `sequence` seperti `list`, `tuple`, `string`, dan `array` atau objek yang dapat diiterasi lainnya.

**For Loop Syntax**

```python
for val in sequence:
    loop body
```

**For Loop Flowchart**

gambar perulangan 1.jpeg

**Example : Python for Loop**

```python
#Program untuk menghitung jumlah semua angka dalam sebuah list
L1 = [2,4,5,7,9]
jumlah = 0

#iterasi di dalam list
for i in L1:
  jumlah = jumlah + i

print(jumlah)
```

##The range function

```python
#Program untuk melakukan iterasi di dalam list melalui indeks
minuman = ["teh", "boba", "kopi", "susu"]
p = len(minuman)

#iterasi di dalam list melalui indeks
for i in range(p):
  print(minuman[i])
```

##for loop with else

```python
#contoh for loop mengggunakan else
for i in range(p):
  print(minuman[i])
else:
  print("Elemen sudah habis")
```

#**Latihan 1**

Diberikan `list_tagihan`. Kamu diminta untuk menghitung jumlah seluruh tagihan!. Gunakan for untuk menyelesaikan tugas ini!

simpan hasilnya pada variabel `total_tagihan`!

```python
list_tagihan = [50000, 75000, -150000, 125000, 300000, -50000, 200000]
total_tagihan = 0

for i in list_tagihan:
  total_tagihan = total_tagihan + i

print("Total tagihannya adalah ", total_tagihan)
```

#**While Loop**

While loop pada python digunakan untuk melakukaniterasi pada blok kode selama nilai dari `text expression` adalah `true`.

**Syntax of while Loop in Python**

```python
while test_expression:
  Body of while
```

**While Loop Flowchart**

gambar while loop.jpeg

Example : Python while Loop

```python
#Program untuk menjmlahkan semua
#bilangan asli hingga suku ke n
#jum=1+2+3...+n

n=10
jumlah = 0
i =1

while i<=n:
  jumlah=jumlah+i
  i=i+1

print(jumlah)
```

##While loop with else

Sama dengan `for loop` kita juga dapat mengkombinasikan `while loop` dengan blok `else`.

Bagian `else` akan dieksekusi jika nilai pada `test expression` pada `while loop` bernilai `false`. Perhatikan contoh berikut :

```python
jum = 0
i = 1
while i <= n:
  jum = jum + i
  i = i + 1
else:
  print(jum) # dieksekusi diluar looping
```

#**Latihan 2**


Diberikan list `tagihan` hitunglah `total_tagihan` menggunakan `while loop`

```python
tagihan = [50000, 75000, 125000, 300000, 200000]

i = 0
jml_tagihan = len(tagihan)
total_tagihan = 0

while i < jml_tagihan:
  total_tagihan = total_tagihan + tagihan[i]
  i = i + 1

print("Total tagihannya adalah ", total_tagihan)
```
